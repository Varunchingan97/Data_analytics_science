```python
def func(x):
    if x<30:
        return 'Age<30'
    elif x<40:
        return '30<Age<40'
    else:
        return 'Age>=40'
```

```python
df['Age_cat']=df['Age'].apply(func)
df
```

```python
def func2(x):
    if x<=94:
        return 'Low'
    elif x<=114:
        return 'Medium'
    else:
        return 'High'
    
#values are taken from 50% and 75% values
```

```python
df['Miles_per_week']=df['Miles'].apply(func2)
df
```

```python
def func3(x):
    if x<=50596.5:
        return 'Lower class'
    elif x<=58668:
        return 'Middle class'
    else:
        return 'Upper class'
    
#values are taken from 50% and 75% values
```

```python
df['Income level']=df['Income'].apply(func3)
df
```

```python
def func4(x):
    if x<=16:
        return 'Grad or less'
    else:
        return 'PG or higher'
    
#Assumed that 16 years of education means that that the customer is graduated
```

```python
df['Edu_level']=df['Education'].apply(func4)
df
```

```python
from statistics import mode
```

```python
df2=pd.crosstab([df['MaritalStatus'],df['Gender'],df['Age_cat'],df['Edu_level']],[df['Income level'], df['Miles_per_week'],df['Fitness']], values='Product', aggfunc='count')
df2
```

Above are the number of purchases that happened per each combinations. This can be used as customer profiling. Most of the values are NaN because we only have 180 records here.

```python
df3=pd.pivot_table(data=df, index=['MaritalStatus','Gender','Age_cat','Edu_level'], columns=['Income level', 'Miles_per_week','Fitness'], values='Product', aggfunc=mode)
df3
```

Above matrix can be used a tool for recommendation based on all parameters. Here we only have 180 records thats why most of the cells are looking as NaN. The product is chosen as per most bought product for a particular combination of input parameters

```python
df3.loc[('Partnered', 'Female','Age<30','Grad or less'),('Lower class','High',4)]
```

Above is the example of how we can use the matrix for recommendation based on our input parameters which are passed as multi level indices.

```python
price=pd.Series([1500, 1750, 2500])
```

```python
tot=df.groupby('Product')['Income'].count().reset_index().rename(columns={'Income':'Total_sales'})
tot
```

```python
tot['Revenue']=tot['Total_sales']*price
tot
```

```python
plt.pie(tot['Revenue'], labels=tot['Product'], autopct='%1.2f%%')
plt.title('Revenue distribution among all products')
plt.show()
```

Revenue distribution among all 3 products are as shown above. Revenues are decreasing in the order KP281, KP481, KP781.
We can say that the revenue distribution is actually good here. Despite of KP781 being much costly compared to other products, it still managed to make comparable revenue.
